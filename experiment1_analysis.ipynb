```python
import pandas as pd

df = pd.read_excel("/workspaces/ASD-Diagnostic_Research_2023-24/Test Data Table.xlsx")
df.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# Analyzing model performance based on ML Algorithm
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Sensitivity', y='Specificity', hue='ML Algorithm', data=df, palette='rainbow')
plt.title('Model Performance by ML Algorithm: Sensitivity vs. Specificity')
plt.xlabel('Sensitivity (Recall)')
plt.ylabel('Specificity')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
# Melt the DataFrame to reshape it
df_melted_algorithm = pd.melt(df, id_vars=['ML Algorithm'], value_vars=['Sensitivity', 'Specificity', 'Precision', 'F1 Score'])
sns.barplot(x='ML Algorithm', y='value', hue='variable', data=df_melted_algorithm, errorbar=None, palette='pastel')
plt.title('Model Performance by ML Algorithm: Sensitivity, Specificity, Precision, F1 Score')
plt.xlabel('Algorithm')
plt.ylabel('Metric Value')
plt.legend(title='Metric')
plt.show()
```

```python
# ANALYZING PERFORMANCE OF SVM MODELS:

df_svm = df[df['ML Algorithm'] == 'SVM'] # Extracting models using SVM algorithm only
df_svm_melted = pd.melt(df_svm, id_vars=['Strategy'], value_vars=['Sensitivity', 'Specificity', 'Precision', 'F1 Score'])

# Scatter plot of performance of SVM models, color coded based on derivative used
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Sensitivity', y='Specificity', hue='Derivative', data=df_svm, palette='muted')
plt.title('Performance of SVM Models by ROI/Derivative: Sensitivity vs. Specificity')
plt.xlabel('Sensitivity (Recall)')
plt.ylabel('Specificity')
plt.legend(title='Derivative')
plt.show()

# Bar plot of performance of SVM models for each strategy
plt.figure(figsize=(12, 8))
sns.barplot(x='Strategy', y='value', hue='variable', data=df_svm_melted, errorbar=None, palette='muted')
plt.title('Performance of SVM Models by Strategy: Sensitivity, Specificity, Precision, F1 Score')
plt.xlabel('Strategy')
plt.ylabel('Metric Value')
plt.legend(title='Metric')
plt.show()
```

```python
# Set df_heatmap as the heatmap DataFrame with ML Algorithms, Derivatives, and Metrics as columns
df_heatmap = pd.melt(df, id_vars=['ML Algorithm', 'Derivative'], value_vars=['Sensitivity', 'Specificity'])
df_heatmap = df_heatmap.pivot_table(index='ML Algorithm', columns='Derivative', values='value', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(df_heatmap, cmap='YlGnBu', annot=True, fmt=".3f", linewidths=.5)
plt.title('Heatmap of Overall Model Performance by Sensitivity & Specificity: ML Algorithm & ROI/Derivative')
plt.show()

# Set df_heatmap as the heatmap DataFrame with ML Algorithms, Derivatives, and Metrics as columns
df_heatmap2 = pd.melt(df, id_vars=['ML Algorithm', 'Derivative', 'Strategy'], value_vars=['Sensitivity', 'Specificity'])
df_heatmap2 = df_heatmap2.pivot_table(index='ML Algorithm', columns=['Derivative', 'Strategy'], values='value')

plt.figure(figsize=(12, 8))
sns.heatmap(df_heatmap2, cmap='YlGnBu', annot=True, fmt=".3f", linewidths=.5)
plt.title('Heatmap of Overall Model Performance by Sensitivity & Specificity: All Variables')
plt.show()
```
